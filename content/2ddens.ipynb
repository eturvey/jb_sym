Be sure to import Symbulate using the following commands.

```python
from symbulate import *
%matplotlib inline
```

For two continuous random variables, plot type `density` uses the simulated $(x, y)$ to estimate the joint probability density function and plot it.

*Example.* Two-dimensional density plot of simulated values of $X$ and $Y$ where $Y$ and $Z$ are i.i.d. Normal(0, 1) and $X = Y + Z$.  Compare with the scatterplot and two-dimensional histogram we explored earlier.  (The random variables $X$ and $Y$ have a Bivariate Normal distribution.)  

```python
Y, Z = RV(Normal(mean=0, sd=1) ** 2)
X = Y + Z
(X & Y).sim(10000).plot(type="density")
```

Any two continuous variables can be visualized this way. Below we look at an Exponential RV with parameter 1 and a Uniform RV distributed over (5,8).

```python
M, N = RV(Exponential(1)*Uniform(5,8))
(M&N).sim(10000).plot(type="density")
```
