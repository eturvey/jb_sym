# Continuous Time Function

Be sure to import Symbulate using the following commands.

```python
from symbulate import *
%matplotlib inline
```

Instead of using `RandomProcess`, we can use `ContinuousTimeFunction` in combination with `RV` in Symbulate for RP's on a continuous set of values. 

This function creates an object represeting continuous time for the RP to work over. From here we can manipulate RV's according to time values.

```python
P = Exponential(1)
t = ContinuousTimeFunction()
Y = RV(P)
X = Y * cos(2 * pi * t)
X.sim(10).plot()
```
