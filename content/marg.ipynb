# Marginal Plot

Be sure to import Symbulate using the following commands.

```python
from symbulate import *
%matplotlib inline
```

We are also able to examine marginal plots of variables in addition to the joint plot between two RVs. We can do this with discrete and random variables, and for many different types of plots. Below are several examples putting this in action. 

```python
die = list(range(1, 6+1, 1)) # this is just a list of the number 1 through 6
rolls = BoxModel(die, size = 5)
X = RV(rolls, min)
Y = RV(rolls, max)
(X&Y).sim(1000).plot(type=["tile","marginal"])
```

```python
Y, Z = RV(Normal(mean=0, sd=1) ** 2)
X = Y + Z
(X & Y).sim(10000).plot(type=["density","marginal"])
```

```python
(X & Y).sim(10000).plot(type=["scatter","marginal"])
```
