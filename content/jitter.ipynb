# Jitter Plots

Be sure to import Symbulate using the following commands.
<a id='rv'></a>

```python
from symbulate import *
%matplotlib inline
```

For discrete random variables it is recommended to use the `jitter=True` option to better visualize coincident points.

*Example.* Roll a die 5 times and look at `X` - the min of these rolls and `Y` - the max of these rolls. Repeat 1000 times and plot this relationship. Note that we examined these variables in the scatterplot section but it was less obvious where higher density areas lay.

```python
die = list(range(1, 6+1, 1)) # this is just a list of the number 1 through 6
rolls = BoxModel(die, size = 5)
X = RV(rolls, min)
Y = RV(rolls, max)
(X&Y).sim(1000).plot(jitter=True)
```

*Example.* Scatterplot of simulated values of $X$ and $Y$ where $Y$ and $Z$ are i.i.d. Poisson(1) and $X = Y + Z$.

```python
Y, Z = RV(Poisson(1) ** 2)
X = Y + Z
(X & Y).sim(1000).plot(jitter=True)
```
