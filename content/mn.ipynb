# Multinomial Distribution

Be sure to import Symbulate using the following commands.

```python
from symbulate import *
%matplotlib inline
```

A **Multinomial** distribution is a generalized form of the Binomial distribution. We want some event to occur `n` times and we have a vector of probabilities, `p` that tell us how likely the event is to occur at that instance. The vector of probabilities must sum to 1. 

The pmf looks like this:
$$
f(x_1,\cdots, x_n) = \frac{n!}{x_1!\cdots x_k!} p_1^{x_1} \cdots p_k^{x_k}
$$


```python
X = RV(Multinomial(n=5,p=[.5,.4,.1]))
X.sim(100)
```

```python
X.sim(100).plot()
```

When we plot these simulations we are given the index along the x-axis (i.e. when the event occured) and the number of occurences at that instance along the y-axis.
