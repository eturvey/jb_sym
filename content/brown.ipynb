# Brownian Motion

Be sure to import Symbulate using the following commands.

```python
from symbulate import *
%matplotlib inline
```

**Brownian Motion**, defined using `BrownianMotionProbabilitySpace` in Symbulate, is a continuous time, continuous state process. A basic instnace of brownian motion at any given time will follow a normal distribution with mean 0 and varaiance that is equal to time.

In Symbulate we can define the `drift` - how the process changes over time (set equal to 1 means that the process increases by an average of 1 for each 1 unit increase), and `scale` - a scaling value that changes magnitude but not behaviour of values in a process.

```python
P = BrownianMotionProbabilitySpace(drift=0, scale=1)
W = RV(P)

w = W.sim(100)
w.plot(tmax=1000)
```

Note that if we look at time 100 we are given a Normal(0,10) RV.

```python
w = W.sim(1000)
w[100].plot()
Normal(0,10).plot()
```

And at time 4 we see a Normal(0,2) RV.

```python
w[4].plot()
Normal(0,2).plot()
```

When we change `drift` to a non-zero value we see how values change over time.

```python
P = BrownianMotionProbabilitySpace(drift=1, scale=1)
W = RV(P)

w = W.sim(100)
w.plot(tmax=10)
```
