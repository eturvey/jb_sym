# Multivariate Normal (MVN) Distribution

Be sure to import Symbulate using the following commands.

```python
from symbulate import *
%matplotlib inline
```

**Multivariate Normal distributions.** A bivariate normal distribution is a special case of a Multivariate Normal distribution (a.k.a. joint Gaussian distribution).  A multivariate normal distribution is specified by
* a mean vector, which specifies the marginal means
* a covariance matrix, which specifies the covariance between each coordinate pair.  (The diagonal entries of the covariance matrix specify the marginal variances

When we simulate values from this distribution each draw is a vector of values. We will then recieve a vector of values is we call the `.mean()` or `.var()` functions on a set of draws, and a matrix of values if we call `.cov()`.

```python
mu = [1, 2, 3]
Sigma = [[1, 1, -2],
         [1, 4, 0],
         [-2, 0, 9]]
X = RV(MultivariateNormal(mean = mu, cov = Sigma))
X.sim(5)
```

```python
x = X.sim(10000)
x.mean(), x.cov()
```

Note that when we try to plot RV's with more than 2 entries we end up with something like we see below. The x-axis values represent the position in the resulting vector the value belongs to, while the y-axis gives us the value itself.

```python
X.sim(100).plot()
```

Alternatively we can use unpacking to break the resulting vector into its parts and examine any of these variables individually or with another variable.

```python
X, Y, Z = RV(MultivariateNormal(mean = mu, cov = Sigma))
(X & Y).sim(10000).plot(alpha = 0.2)
```
